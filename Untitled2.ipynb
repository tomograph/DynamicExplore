```python
import numpy as np
import skimage.transform
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = np.zeros((64,64))
img1 = img.copy()
img1[32,32] = 1
```

```python
plt.imshow(img1, cmap="gray")
```

```python
radon1 = skimage.transform.radon(img1, np.linspace(0,180,200))
```

```python
plt.imshow(radon1, cmap="gray")
```

```python
img2 = img.copy()
img2[20,10] = 1
plt.imshow(img2, cmap = "gray")
```

```python
radon2 = skimage.transform.radon(img2, np.linspace(0,180,200))
plt.imshow(radon2, cmap="gray")
```

```python
img3 = img.copy()
img3[30,40] = 1
plt.imshow(img3, cmap = "gray")
```

```python
radon3 = skimage.transform.radon(img3, np.linspace(0,360,200))
plt.imshow(radon3, cmap="gray")
```
