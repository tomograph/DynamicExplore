```python
import odl


phantom_space = odl.uniform_discr([-50., -50.], [ 50.,  50.], (100, 100))
phantom = odl.phantom.shepp_logan(phantom_space, modified=True)
phantom.show()
```

```python
fft = odl.trafos.WaveletTransform(phantom_space, 'haar')
```

```python
fft_phantom = fft(phantom)
```

```python
fft_phantom.show()
```

```python
ifft = odl.trafos.WaveletTransformInverse(phantom_space, 'db')
```

```python
phantom_back = ifft(fft_phantom)
```

```python
phantom_back.show()
```
